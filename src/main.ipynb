```python
# main.ipynb (요약된 코드 예시)

import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
data = pd.read_csv('../data/sample_data.csv')

# 간단한 통계
print("데이터 요약:")
print(data.describe())

# 시각화
plt.plot(data['time'], data['value'])
plt.title("Sample Data Visualization")
plt.xlabel("Time")
plt.ylabel("Value")
plt.savefig('../results/output_plot.png')
plt.show()
```

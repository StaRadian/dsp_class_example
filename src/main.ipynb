# 데이터·코드 관리 기본 툴 만들기 예제

본 notebook은 샘플 데이터를 불러와 간단히 시각화 하는 예시로, 재현 가능한 코드 관리 환경의 기본구조를 보여줍니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 데이터 파일 위치
GITHUB_RAW_URL = "https://raw.githubusercontent.com/StaRadian/dsp_class_example/main/data/sample_data.csv"
local_path = "../data/sample_data.csv"

# 경로 감지
if os.path.exists(local_path):
    print("Local file found — loading from local path.")
    data = pd.read_csv(local_path)
else:
    print("Local file not found — loading from GitHub URL.")
    data = pd.read_csv(GITHUB_RAW_URL)

# 시각화
plt.plot(data["time"], data["value"], marker="o")
plt.title("Sample Data Visualization")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
